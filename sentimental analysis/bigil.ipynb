```python
def label_sentiment(x:float):
  if x<-0.05:return 'negative'
  if x>0.35:return 'positive'
  return 'neutral'
```

```python
sia=SIA()
```

```python
df['sentiment']=[sia.polarity_scores(x)['compound'] for  x in tqdm(df['tweet'])]
```

```python
df['overall']=df['sentiment'].apply(label_sentiment)
```

```python
non_meaningful_words=['cr','amp','rs','u','l']
def remove_non_meaningful_words(text):
  tokens=text.split()
  filtered=[token for token in tokens if token not in non_meaningful_words]
  filtered_text=' '.join(filtered)
  return filtered_text
df['tweet']=df['tweet'].apply(remove_non_meaningful_words)
```
